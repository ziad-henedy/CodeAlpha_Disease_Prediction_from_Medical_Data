```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

```

```python
df = pd.read_csv("cardio_train.csv")

display(df)
```

## Data Preprocessing

No need for OneHotEncoding since all values are numeric

### Data Separation into x and y

```python
x = df.drop('cardio',axis=1)
y = df['cardio']
```

### Splitting into train and test

```python

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

## Building the Model

### 1st Model RandomForest

### Training the model

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
# what fits does is Build a forest of trees from the training set (x, y).
rf.fit(x_train,y_train)
```

```python
y_pred = rf.predict(x_test)
y_pred
```

```python
compare = pd.DataFrame(columns=['Actual','Prediction'])
""" prediction = pd.DataFrame(y_pred)
actual = pd.DataFrame(y_test)"""

compare['Actual'] = y_test
compare['Prediction'] = y_pred

```

```python
compare['Prediction'] = compare['Prediction'].map({0:'no',1:'yes'})
compare['Actual'] = compare['Actual'].map({0:'no',1:'yes'})
```

## Checking Accuracy and Precision

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score


acc = accuracy_score(y_test,y_pred)
prec = precision_score(y_test,y_pred)

print("Accuracy:", acc*100, "%")
print("Precision:",prec*100,"%")
```
